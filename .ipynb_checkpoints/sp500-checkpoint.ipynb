```python
# Manipulación de datos
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import random

# Modelado tradicional
from sklearn.model_selection import TimeSeriesSplit, cross_val_score, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, BayesianRidge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from xgboost import XGBRegressor

# Métricas de evaluación
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, make_scorer

# Visualización
import matplotlib.pyplot as plt

# Redes neuronales (Keras/Tensorflow)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, SimpleRNN, Conv1D, MaxPooling1D, Flatten, Bidirectional
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.optimizers import Adam


# Control de warnings
import warnings
warnings.filterwarnings("ignore")


from datetime import timedelta


# Semilla fija para reproducibilidad
seed = 42
os.environ['PYTHONHASHSEED'] = str(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)

# Asegurar determinismo (si está disponible)
try:
    tf.keras.utils.set_random_seed(seed)
    tf.config.experimental.enable_op_determinism()
except:
    pass

```

```python

# Cargar el archivo CSV
relative_path = os.path.join("processed_files", "sp500_historico_normalized.csv")
df = pd.read_csv(relative_path, parse_dates=['Date'])
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)

# Asegurar tipos numéricos correctos
for col in ['Close', 'High', 'Low', 'Open', 'Volume']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.dropna(subset=['Close', 'High', 'Low', 'Open', 'Volume'], inplace=True)
df = df.astype({col: 'float32' for col in ['Close', 'High', 'Low', 'Open', 'Volume']})

# Feature engineering: se calculan con información hasta t (inclusive)
df['ma_5'] = df['Close'].rolling(window=5).mean()
df['std_5'] = df['Close'].rolling(window=5).std()
df['momentum_5'] = df['Close'] - df['Close'].shift(5)

# Desplazar **todos** los targets a t+1
df['target_open_t+1']  = df['Open'].shift(-1)
df['target_high_t+1']  = df['High'].shift(-1)
df['target_low_t+1']   = df['Low'].shift(-1)
df['target_close_t+1'] = df['Close'].shift(-1)

# Eliminar nulos generados por rolling y shift
df.dropna(inplace=True)

# Definir variables de entrada y salida
feature_cols = [
    'Open','High','Low','Close','Volume',
    'ma_5','std_5','momentum_5'
]
X = df[feature_cols]
# Diccionario de targets individuales
Ys = {
    'Open':  df['target_open_t+1'],
    'High':  df['target_high_t+1'],
    'Low':   df['target_low_t+1'],
    'Close': df['target_close_t+1']
}


# Separar en train/test conservando orden temporal
train_size = int(len(X) * 0.8)
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
```

# Modelos Tradicionales

```python

# 1) Defino los pipelines base
base_models = {
    'LinearRegression':     LinearRegression(),
    'Ridge':                Ridge(),
    'Lasso':                Lasso(),
    'BayesianRidge':        BayesianRidge(),
    'DecisionTree':         DecisionTreeRegressor(max_depth=5, min_samples_leaf=5, random_state=42),
    'RandomForest':         RandomForestRegressor(n_estimators=100, random_state=42),
    'HistGradientBoosting': HistGradientBoostingRegressor(random_state=42),
    'XGBoost':              XGBRegressor(n_estimators=100, random_state=42, objective='reg:squarederror')
}

# 2) Creamos pipelines con escalado
pipelines = {
    name: Pipeline([('scaler', StandardScaler()), ('model', model)])
    for name, model in base_models.items()
}

# 3) Entreno por cada target
results_trad = []

for target_name, y in Ys.items():
    # split temporal
    y_train = y.iloc[:train_size]
    y_test  = y.iloc[train_size:]
    
    for name, pipe in pipelines.items():
        pipe.fit(X_train, y_train)
        y_pred = pipe.predict(X_test)
        results_trad.append({
            'Target': target_name,
            'Model':  name,
            'MAE':    mean_absolute_error(y_test, y_pred),
            'MSE':    mean_squared_error(y_test, y_pred),
            'RMSE':   np.sqrt(mean_squared_error(y_test, y_pred)),
            'R2':     r2_score(y_test, y_pred)
        })

df_results_trad = pd.DataFrame(results_trad)

# df_results_trad tiene columnas: ['Target','Model','MAE','MSE','RMSE','R2']

for modelo in df_results_trad['Model'].unique():
    print(f"\n🔍 Métricas para el modelo: {modelo}\n")
    # Filtramos sólo las filas de este modelo, ponemos el Target como índice
    df_model = (
        df_results_trad[df_results_trad['Model'] == modelo]
        .set_index('Target')[['MAE','MSE','RMSE','R2']]
        .T     # transponemos para que las métricas queden en filas
        .round(4)
    )
    display(df_model)

print(df_results_trad.set_index('Model').sort_values(by='RMSE'))

```

```python

plt.style.use('ggplot')

# Métricas a graficar
metrics = ['MAE', 'MSE', 'RMSE', 'R2']

# Para cada modelo
for modelo in df_results_trad['Model'].unique():
    df_model = df_results_trad[df_results_trad['Model'] == modelo].set_index('Target')
    
    # Creamos una figura por modelo
    fig, axes = plt.subplots(1, len(metrics), figsize=(20, 4), sharey=False)
    
    for ax, metric in zip(axes, metrics):
        # Valores de la métrica para cada Target
        valores = df_model[metric]
        targets = df_model.index.tolist()
        
        bars = ax.bar(targets, valores, color='skyblue')
        ax.set_title(f'{modelo} — {metric}')
        ax.set_xlabel('Target')
        ax.set_ylabel(metric)
        ax.set_xticklabels(targets, rotation=45, ha='right')
        
        # Etiquetas encima de cada barra
        for bar in bars:
            yval = bar.get_height()
            ax.text(bar.get_x() + bar.get_width() / 2, yval,
                    f'{yval:.2f}' if metric != 'R2' else f'{yval:.4f}',
                    ha='center', va='bottom', fontsize=8)
    
    fig.suptitle(f'Métricas por target para el modelo {modelo}', fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

# Redes Neuoranles

```python
# Reescalado general para entradas
scaler_nn = MinMaxScaler()
X_scaled = scaler_nn.fit_transform(X)
X_train_nn = X_scaled[:train_size]
X_test_nn = X_scaled[train_size:]

# Para modelos secuenciales
X_train_seq = X_train_nn.reshape((X_train_nn.shape[0], X_train_nn.shape[1], 1))
X_test_seq = X_test_nn.reshape((X_test_nn.shape[0], X_test_nn.shape[1], 1))

# Función de evaluación unificada
def evaluar_modelo(model_name, target_name, modelo, X_test, y_test):
    y_pred = modelo.predict(X_test).flatten()
    y_test = y_test.flatten()
    return {
        'Target': target_name,
        'Model': model_name,
        'MAE': mean_absolute_error(y_test, y_pred),
        'MSE': mean_squared_error(y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'R2': r2_score(y_test, y_pred)
    }

# Lista de resultados
resultados_nn = []

# Entrenamiento por cada variable target
for target_name, y in Ys.items():
    # Reescalar target individualmente
    scaler_y = MinMaxScaler()
    y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))
    y_train_nn = y_scaled[:train_size]
    y_test_nn = y_scaled[train_size:]

    # ANN
    model_ann = Sequential([
        Dense(64, activation='relu', input_shape=(X_train_nn.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1)
    ])
    model_ann.compile(optimizer=Adam(), loss='mse')
    model_ann.fit(X_train_nn, y_train_nn, epochs=50, batch_size=32, verbose=0)
    resultados_nn.append(evaluar_modelo("ANN", target_name, model_ann, X_test_nn, y_test_nn))

    # LSTM
    model_lstm = Sequential([
        LSTM(64, input_shape=(X_train_seq.shape[1], 1)),
        Dense(1)
    ])
    model_lstm.compile(optimizer=Adam(), loss='mse')
    model_lstm.fit(X_train_seq, y_train_nn, epochs=50, batch_size=32, verbose=0)
    resultados_nn.append(evaluar_modelo("LSTM", target_name, model_lstm, X_test_seq, y_test_nn))

    # GRU
    model_gru = Sequential([
        GRU(64, input_shape=(X_train_seq.shape[1], 1)),
        Dense(1)
    ])
    model_gru.compile(optimizer=Adam(), loss='mse')
    model_gru.fit(X_train_seq, y_train_nn, epochs=50, batch_size=32, verbose=0)
    resultados_nn.append(evaluar_modelo("GRU", target_name, model_gru, X_test_seq, y_test_nn))

    # BiLSTM
    model_bilstm = Sequential([
        Bidirectional(LSTM(64), input_shape=(X_train_seq.shape[1], 1)),
        Dense(1)
    ])
    model_bilstm.compile(optimizer=Adam(), loss='mse')
    model_bilstm.fit(X_train_seq, y_train_nn, epochs=50, batch_size=32, verbose=0)
    resultados_nn.append(evaluar_modelo("BiLSTM", target_name, model_bilstm, X_test_seq, y_test_nn))

    # CNN + LSTM
    model_cnn_lstm = Sequential([
        Conv1D(64, kernel_size=2, activation='relu', input_shape=(X_train_seq.shape[1], 1)),
        MaxPooling1D(pool_size=2),
        LSTM(50),
        Dense(1)
    ])
    model_cnn_lstm.compile(optimizer=Adam(), loss='mse')
    model_cnn_lstm.fit(X_train_seq, y_train_nn, epochs=50, batch_size=32, verbose=0)
    resultados_nn.append(evaluar_modelo("CNN+LSTM", target_name, model_cnn_lstm, X_test_seq, y_test_nn))

    # SimpleRNN
    model_simplernn = Sequential([
        SimpleRNN(64, input_shape=(X_train_seq.shape[1], 1)),
        Dense(1)
    ])
    model_simplernn.compile(optimizer=Adam(), loss='mse')
    model_simplernn.fit(X_train_seq, y_train_nn, epochs=50, batch_size=32, verbose=0)
    resultados_nn.append(evaluar_modelo("SimpleRNN", target_name, model_simplernn, X_test_seq, y_test_nn))

# Convertir a DataFrame
df_results_nn = pd.DataFrame(resultados_nn)

# Mostrar resultados igual que en modelos tradicionales
for modelo in df_results_nn['Model'].unique():
    print(f"\n🔍 Métricas para el modelo: {modelo}\n")
    df_model = (
        df_results_nn[df_results_nn['Model'] == modelo]
        .set_index('Target')[['MAE','MSE','RMSE','R2']]
        .round(4)
    )
    display(df_model)

# Orden general por RMSE
print(df_results_nn.sort_values(by='RMSE').set_index('Model'))
```

```python
plt.style.use('ggplot')

# Métricas a graficar
metrics = ['MAE', 'MSE', 'RMSE', 'R2']

# Para cada modelo de red neuronal
for modelo in df_results_nn['Model'].unique():
    df_model = df_results_nn[df_results_nn['Model'] == modelo].set_index('Target')
    
    # Creamos una figura por modelo
    fig, axes = plt.subplots(1, len(metrics), figsize=(20, 4), sharey=False)
    
    for ax, metric in zip(axes, metrics):
        # Valores de la métrica para cada Target
        valores = df_model[metric]
        targets = df_model.index.tolist()
        
        bars = ax.bar(targets, valores, color='salmon')  # color diferente para diferenciar redes
        ax.set_title(f'{modelo} — {metric}')
        ax.set_xlabel('Target')
        ax.set_ylabel(metric)
        ax.set_xticklabels(targets, rotation=45, ha='right')
        
        # Etiquetas encima de cada barra
        for bar in bars:
            yval = bar.get_height()
            ax.text(bar.get_x() + bar.get_width() / 2, yval,
                    f'{yval:.2f}' if metric != 'R2' else f'{yval:.4f}',
                    ha='center', va='bottom', fontsize=8)
    
    fig.suptitle(f'Métricas por target para el modelo {modelo}', fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
# Añadir columna 'Tipo' en cada DataFrame antes de unir
df_results_trad['Tipo'] = 'Tradicional'
df_results_nn['Tipo'] = 'Red Neuronal'

# Unir ambos DataFrames
df_resultados_combinado = pd.concat([df_results_trad, df_results_nn], ignore_index=True)

# Ordenar por RMSE si lo deseas
df_resultados_combinado = df_resultados_combinado.sort_values(by='RMSE')

# Mostrar los primeros resultados combinados
print(df_resultados_combinado.head())

# guardar en CSV para Power BI u otros análisis
df_resultados_combinado.to_csv("resultados_modelos_combinados.csv", index=False)
```
